```python
#Librerias utilizadas en el trabajo de datos y predicci贸n
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sbn
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import Dense,Flatten
from datetime import timedelta
from pptx.util import Inches
from pptx import Presentation
```

```python
#Importacion del dataset que contiene los computadores, con precio, id, tienda y fecha
df = pd.read_excel("precioFechaPorId.xlsx")#, index_col=9, parse_dates=True#)
```

```python
#Eliminaci贸n de columna Unnamed: 0 que se crea al importar el xlsx
try:
    df = df.drop({"Unnamed: 0"}, axis=1)
except:
    x=2
```

```python
df = df.drop_duplicates(["ID_SKU", "FECHA", "TIENDA"], keep='first')
```

```python
#Funci贸n que retorna las tiendas que pertenecen a un sku en especifico
def BuscarTiendas1(id):
    final = df[df["ID_SKU"]== id]    
    return final
```

```python
#Funci贸n que retorna las tiendas que pertenecen a un sku en especifico
def BuscarTiendas(id):
    final = df[df["ID_SKU"]== id]
    tiendas = final.groupby(["TIENDA"]).count().index
    return tiendas
```

```python
j = int(input("Ingrese id_sku del producto: "))
final = BuscarTiendas1(j)
tienda = BuscarTiendas(j)
```
